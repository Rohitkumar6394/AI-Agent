```python
entity_memory = {}
```

```python
def save_entity(user_text):

    text = user_text.lower()

    if "my name is" in text:
        name = user_text.split("my name is")[-1].strip()
        entity_memory["name"] = name

    elif "i study at" in text:
        college = user_text.split("I study at")[-1].strip()
        entity_memory["college"] = college

    elif "i live in" in text:
        city = user_text.split("I live in")[-1].strip()
        entity_memory["city"] = city
```

```python
MEMORY_FILE = "entity_memory.txt"

def save_to_file():

    with open(MEMORY_FILE, "w") as f:

        for key, value in entity_memory.items():
            f.write(f"{key}:{value}\n")
```

```python
def load_memory():

    global entity_memory

    entity_memory = {}

    try:

        with open(MEMORY_FILE, "r") as f:

            for line in f:

                key, value = line.strip().split(":",1)

                entity_memory[key] = value

    except:
        pass
```

```python
def entity_agent(user_input):

    load_memory()

    save_entity(user_input)

    save_to_file()

    if "what is my name" in user_input.lower():

        return entity_memory.get(
            "name",
            "I don't know your name yet."
        )

    if "where do i live" in user_input.lower():

        return entity_memory.get(
            "city",
            "I don't know your city yet."
        )

    return "Entity stored successfully."
```

```python
while True:

    user = input("You: ")

    if user.lower() == "exit":
        break

    print(
        "Agent:",
        entity_agent(user)
    )
```
