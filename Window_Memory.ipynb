```python
from transformers import pipeline

agent = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    device_map="auto"
)

MEMORY_FILE = "window_memory.txt"
WINDOW_SIZE = 5

open(MEMORY_FILE, "a").close()

print("Agent Ready")
```

```python
def get_window_memory():

    with open(MEMORY_FILE, "r", encoding="utf-8") as f:
        lines = f.readlines()

    return "".join(lines[-WINDOW_SIZE:])


def save_memory(text):

    with open(MEMORY_FILE, "a", encoding="utf-8") as f:
        f.write(text + "\n")
```

```python
def chat(user_input):

    memory = get_window_memory()

    prompt = f"""
Previous Conversation:
{memory}

User: {user_input}
Assistant:
"""

    result = agent(
        prompt,
        max_new_tokens=80,
        return_full_text=False
    )

    answer = result[0]["generated_text"]

    save_memory(f"User: {user_input}")
    save_memory(f"Assistant: {answer}")

    return answer
```

```python
while True:

    msg = input("You: ")

    if msg.lower() == "exit":
        break

    print("Agent:", chat(msg))
```
